```python
import pandas as pd
```

```python
faang = pd.DataFrame()

for ticker in ['AAPL', 'AMZN', 'FB', 'GOOG', 'NFLX']:
    df = pd.read_csv(f'../ch03/data/{ticker.lower()}.csv')
    df.loc[:, "ticker"] = ticker
    
    faang = pd.concat([faang, df])
    
faang.to_csv('faang.csv', index=False)
```

```python
faang = faang.assign(
    date=lambda x: pd.to_datetime(x['date']),
    volume=lambda x: x['volume'].astype(pd.Int64Dtype())
).sort_values(by=["date", "ticker"])

faang.info()
```

```python
faang.nsmallest(7, columns='volume')
```

```python
melted_faang = faang.melt(id_vars=['date', 'ticker'], value_vars=['open', 'high', 'low', 'close', 'volume'])

melted_faang.head()
```
